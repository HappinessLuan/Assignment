```python
def check_zander():
    length = float(input("Enter the length of the zander in centimeters: "))
    limit = 42
    if length < limit:
        difference = limit - length
        print(f"Please release the fish back into the lake. It is {difference:.1f} cm below the size limit.")
    else:
        print("The fish meets the size limit.")
if __name__ == "__main__":
    check_zander()
```
